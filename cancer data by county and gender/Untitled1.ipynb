```python
from bs4 import BeautifulSoup
import re
soup = BeautifulSoup(open("Greece_Affinio_HTML.html"), "html.parser")
import pandas as pd

num_imgs_list = []
for img_number in soup.find_all(class_="info done"):
    print(img_number.contents)
    num_imgs_list.append(img_number.contents[0].split()[0])

list_tags = []
for list_tag in soup.find_all(class_="labels done"):
    soup_nested = BeautifulSoup(list_tag.getText(),'lxml')
    tag_list = []
    for tag in list_tag.findAll('li'):
#         print(tag.contents)
        tag_list.append(str(tag.string).strip())
    list_tags.append(tag_list)

img_list = []
for img_tag in soup.find_all('img', {'src': re.compile(r'.*?\.jpg')}):
    soup_nested = BeautifulSoup(img_tag.getText(),'lxml')
    img_list.append(img_tag.get('src'))

df = pd.DataFrame({'img_count' : num_imgs_list, 'img_tags':list_tags, 'img_link' : img_list})
print(df)
df.to_csv('affinio_data_breakdown_greece.csv', sep=',', encoding='utf-8')
```
